```python
#Calculate the mean, median, and standard deviation for the Score column.
mean=df1['score'].mean()
median=df1['score'].median()
std=df1['score'].std()
print(f"Mean is:{mean}")
print(f"Median is:{median}")
print(f"Std is:{std}")
```

```python
#Identify the country with the highest and lowest happiness scores.
highest_score = df1['score'].idxmax()
lowest_score = df1['score'].idxmin()
highest_score_country = df1.loc[highest_score, 'Country name']
lowest_score_country = df1.loc[lowest_score, 'Country name']
print(f"Highest score is from: {highest_score_country}")
print(f"Lowest score is from: {lowest_score_country}")
```

```python
#For the filtered dataset - Sort the dataset by GDP per Capita in descending order and display the top 10 rows.
sorted_df=df1.sort_values(by='Log GDP per capita',ascending=False)
sorted_df
```

```python
#Create a new column called Happiness Category that categorizes countries into three categories based on their Score:
def categorize_Score(score):
    if (score < 4):
        return "Low"
    elif (4 <= score <= 6):
        return "medium"
    elif (score > 6):
        return "high"
df1['Happiness Category'] = df1['score'].apply(categorize_Score)
df1
```

```python
#Plot the top 10 happiest countries by Score using a bar chart.
Top_10=df1.sort_values(by='score',ascending=False).head(10)
Top_10
bar = Top_10.plot(kind='bar',
             x='Country name',
             y='score',
             color='skyblue',
             title='Top 10 happiest country')
bar.set_xlabel('Country name')
bar.set_ylabel('score')
bar.set_ylim(0, 10)
```

```python
#Plot the top 10 unhappiest countries by Score using a Line chart.
Top_10_unhappy = df1.sort_values(by='score', ascending=True).head(10)

plt.plot(Top_10_unhappy['Country name'], Top_10_unhappy['score'], marker='o')
plt.xticks(ticks=range(len(Top_10_unhappy)), labels=Top_10_unhappy['Country name'], rotation=45)

plt.xlabel('Country Name')
plt.ylabel('Score')
plt.title('Top 10 Unhappiest Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#for the Score column to show its distribution and also interpret.
plt.figure(figsize=(8, 6))
plt.hist(df1['score'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Happiness Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Happiness Scores', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Define the countries in South Asia with a list
south_asian_countries = [
    "Afghanistan",
    "Bangladesh",
    "Bhutan",
    "India",
    "Maldives",
    "Nepal",
    "Pakistan",
    "Sri Lanka"
]
South_asian=df1[df1['Country name'].isin(south_asian_countries)]
South_asian
#Save the filtered dataframe as separate CSV files for future use.
South_asian.to_csv('South_asian.csv', index=False)
```

```python
df=pd.read_csv("/content/South_asian.csv")
df
```

```python
#Using the SouthAsia DataFrame, create a new column called Composite Score
South_asian['Composite score']=(0.40*South_asian['Log GDP per capita'])+(0.30*South_asian['Social support'])+(0.30*South_asian['Healthy life expectancy'])
South_asian
```

```python
#Rank the South Asian countries based on the Composite Score in descending order.
South_asian['Rank']=South_asian['Composite score'].rank(ascending=False)
South_asian=South_asian.sort_values(by='Rank')
South_asian
```

```python
#Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.
Top_5=South_asian.head(5)
ax = Top_5.plot(kind='bar',
             x='Country name',
             y='Composite score',
             color='skyblue',
             title='Compsoite Score')
ax.set_xlabel('Country name')
ax.set_ylabel('score')
```

```python
#Identify outlier countries in South Asia based on their Score and GDP per Capita.
Q1_score = South_asian['score'].quantile(0.25)
Q3_score = South_asian['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

Q1_gdp = South_asian['Log GDP per capita'].quantile(0.25)
Q3_gdp = South_asian['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers = South_asian[
    (South_asian['score'] < lower_bound_score) | (South_asian['score'] > upper_bound_score) |
    (South_asian['Log GDP per capita'] < lower_bound_gdp) | (South_asian['Log GDP per capita'] > upper_bound_gdp)
]
#Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
outliers
df.plot.scatter(x='Log GDP per capita', y='score', title='Outliers')
```

```python
#Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation {pearson correlation} with the Score for South Asian countries.
free=South_asian['Freedom to make life choices'].corr(South_asian['score'])
genoricity=South_asian['Generosity'].corr(South_asian['score'])
print(f"Correlation between Freedom to make life choices and score is:{free}")
print(f"Correlation between Generosity and score is:{genoricity}")
df.plot.scatter(x='Freedom to make life choices', y='score', title='Freedom to make life choice', color='skyblue')
df.plot.scatter(x='Generosity', y='score', title='Generosity', color='green')
```

```python
#Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score for each South Asian country.
South_asian['GDP-Score Gap']=South_asian['Log GDP per capita']-South_asian['score']
South_asian
```

```python
#Identify outlier countries in both regions based on Score and GDP per Capita.
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

south_asia_outliers_score = detect_outliers(South_asian, 'score')
south_asia_outliers_gdp = detect_outliers(South_asian, 'Log GDP per capita')

middle_east_outliers_score = detect_outliers(middle_east, 'score')
middle_east_outliers_gdp = detect_outliers(middle_east, 'Log GDP per capita')

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.scatter(South_asian['Log GDP per capita'], South_asian['score'], color='blue', label='Non-Outliers')
plt.scatter(south_asia_outliers_score['Log GDP per capita'], south_asia_outliers_score['score'], color='red', label='Outliers (Score)')
plt.scatter(south_asia_outliers_gdp['Log GDP per capita'], south_asia_outliers_gdp['score'], color='green', label='Outliers (GDP per Capita)')
plt.title("South Asia: Outliers ")
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.scatter(middle_east['Log GDP per capita'], middle_east['score'], color='blue', label='Non-Outliers')
plt.scatter(middle_east_outliers_score['Log GDP per capita'], middle_east_outliers_score['score'], color='red', label='Outliers (Score)')
plt.scatter(middle_east_outliers_gdp['Log GDP per capita'], middle_east_outliers_gdp['score'], color='green', label='Outliers (GDP per Capita)')
plt.title("Middle East: Outliers ")
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

print("South Asia Outliers ")
print(south_asia_outliers_score[['Country name', 'score', 'Log GDP per capita']])
print(south_asia_outliers_gdp[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Outliers ")
print(middle_east_outliers_score[['Country name', 'score', 'Log GDP per capita']])
print(middle_east_outliers_gdp[['Country name', 'score', 'Log GDP per capita']])
```
